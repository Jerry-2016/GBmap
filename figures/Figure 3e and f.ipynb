```r
suppressPackageStartupMessages({
    library(ggplot2)
    library(Seurat)
    library(dittoSeq)
})
```

```r
gbm <- readRDS('data/extended_GBmap.rds')
gbm
```

```r
options(repr.plot.height = 10, repr.plot.width = 10)
dittoDimPlot(gbm, do.raster = TRUE,
             reduction.use = "umap", var = "author",
             color.panel = rep('#e5e5e5', 30),
             do.label = F, labels.size = 3, size = 1)
```

```r
ggsave('figures/extended_GBmap_grey_background.pdf', height = 10,
       width = 12)
```

```r
gbm_subset <- subset(gbm, subset = gbmap == 'core')
```

```r
options(repr.plot.height = 10, repr.plot.width = 10)

dittoDimPlot(gbm_subset, 
             reduction.use = "umap", var = "high_hierarchy", 
            do.raster = TRUE,
             color.panel = c('#d62728', '#ff7f0e', '#3182bd','#31a354',
                '#e7298a','#e6ab02', '#bf5b17','#00F6B3','#bcbd22',
                '#393b79','#637939','#8c6d31','#636363','#8c564b',
                '#7b4173', '#7fc97f','#beaed4','#fdc086',
                '#756bb1','#17becf'),
            size = 0.5) + NoAxes() 
```

```r
ggsave('figures/extended_GBmap_only_core_reference.pdf', height = 10,
       width = 12)
```

```r
# Neutrophil markers

options(repr.plot.height = 15, repr.plot.width = 20)
ggrastr::rasterize(Nebulosa::plot_density(gbm, c('FCGR3B', 'CSF3R', 'CXCR2','G0S2',
                        'IL1R2', 'FPR2', 'CXCL1'),
                      reduction = 'umap',
                                         size=0.2), dpi=300)
```
