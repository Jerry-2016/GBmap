```r
suppressPackageStartupMessages({
    library(ggplot2)
    library(Seurat)
    library(dplyr)
    library(dittoSeq)
})
```

```r
tils <- readRDS('/data/subset_tils_Ruiz2021_dataset.rds')
```

```r
options(repr.plot.height = 5, repr.plot.width = 5)
dittoDimPlot(tils, size = 3,
             reduction.use = "umap", var = "ident", 
             do.raster = TRUE)
```

```r
ggsave('figures/Ruiz2021_tils.pdf')
```

```r
options(repr.plot.height = 6, repr.plot.width = 8)
tils@misc$markers %>%
    group_by(cluster) %>%
    top_n(n = 10, wt = avg_log2FC) -> top10
DoHeatmap(tils, features = top10$gene,
         group.colors = dittoColors()) + scale_fill_viridis_c()
```

```r
ggsave('figures/Ruiz2021_tils_heatmap.pdf')
```
