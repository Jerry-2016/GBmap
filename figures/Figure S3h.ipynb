```r
suppressPackageStartupMessages({
    library(ggplot2)
    library(Seurat)
    library(dittoSeq)
})
```

```r
reference <- readRDS('data/core_GBmap.rds')
```

```r
# to be able to organize it in a decreasing order in the plot
reference$Tissue = recode(reference$Tissue, 
                           nan = '1. NA',
                         Normal = "5. Periphery",
                     Periphery = '5. Periphery',
                     Tumor = '4. Tumor',
                           )
```

```r
dittoDimPlot(reference, do.raster = TRUE,
             reduction.use = "umap", var = "Tissue", order = 'decreasing',
              color.panel = c(rev(dittoColors()[5:6]), '#e5e5e5'),
             do.label = F, labels.size = 3, size = 0.5)
```

```r
ggsave('figures/core_GBmap_multisector_biopsy_studies.pdf', width = 7.5)
```
