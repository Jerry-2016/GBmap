```python
sc.set_figure_params(dpi=100, dpi_save = 300, color_map = 'viridis_r')
sc.settings.verbosity = 1
sc.logging.print_header()
pl.rcParams['pdf.fonttype'] = 42
sc.settings.figdir = 'figures'
```

```python
adata = sc.read('data/core_GBmap.h5ad')
```

### Visualization of marker genes using scanpy

```python
sc.tl.rank_genes_groups(adata, 'annotation_level_3', method='wilcoxon')
sc.pl.rank_genes_groups(adata, n_genes=35, sharey=False)
```

```python
sc.pl.rank_genes_groups_dotplot(adata, n_genes=5, values_to_plot='logfoldchanges', 
                                min_logfoldchange=2, vmax=5, vmin=-5, cmap='BrBG',
                               save = 'figures/GBmap_dotplot_annotation_level_3.pdf')
```
