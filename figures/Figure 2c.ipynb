```r
suppressPackageStartupMessages({
    library(ggplot2)
    library(data.table)
    library(Seurat)
    library(dplyr)
    library(tidyr)
    library(dittoSeq)
    library(flipPlots)
    set.seed(123)
})
```

```r
sessionInfo()
```

```r
reference <- readRDS('data/core_GBmap.rds')
```

```r
options(repr.plot.height = 12, repr.plot.width = 12)
multi_dittoDimPlot(reference, 
                   # order = 'randomize',
             reduction.use = "umap", var = c('annotation_level_1', 'annotation_level_2',
                                             'annotation_level_3', 'annotation_level_4'),
            do.raster = TRUE,
            size = 0.5) + NoAxes() 
```

### Sankey plot

#### Level 1 to 2

```r
classTab <- data.frame(row.names = colnames(reference), 
                       detailed_level = reference@meta.data$annotation_level_2, 
                       coarse_level = reference@meta.data$annotation_level_1)
classTab
```

```r
detailed<- classTab$detailed_level
coarse <- classTab$coarse_level

table(detailed, coarse)
chisq.test(table(detailed, coarse))
```

```r
transMat <- as.data.frame(table(coarse, detailed), stringsAsFactors = F)
transMat$detailed<- factor(transMat$detailed, levels = names(table(reference@meta.data$annotation_level_2)))
transMat$coarse <- factor(transMat$coarse, levels = names(table(reference@meta.data$annotation_level_1)))
```

```r
table(classTab$detailed_level)
```

```r
transMat <- as.data.frame(table(coarse, detailed), stringsAsFactors = F)
transMat$coarse <- factor(transMat$coarse, 
                         levels = c('Neoplastic', 'Non-Neoplastic'))
transMat$detailed<- factor(transMat$detailed, 
                             levels = c('Differentiated-like',
                                        'Stem-like',
                                        'Glial-Neuronal',
                                        'Myeloid',
                                        'Lymphoid',
                                        'Vascular')
                            )
```

```r
widget = SankeyDiagram(transMat[, -3],
              max.categories = 60,
              link.color = "Source", 
              label.show.varname = TRUE,
              font.size = 0,
              weights = transMat$Freq,
              node.width = 100,
              colors = c('#dc8691', '#78c59e'),
              node.position.automatic = FALSE
) 
widget
```

```r
htmlwidgets::saveWidget(widget, 'figures/sankey_plot_from_level_1_to_level_2.html')
```

```r
webshot::webshot('figures/sankey_plot_from_level_1_to_level_2.html', 
                 file = 'figures/sankey_plot_from_level_1_to_level_2.pdf')
```
