R environment

```r
suppressPackageStartupMessages({
    library(ggplot2)
    library(Seurat)
    library(dplyr)
    library(dittoSeq)
})
```

```r
reference <- readRDS('data/core_GBmap.rds')
```

```r
options(repr.plot.height = 7, repr.plot.width = 8)
dittoDimPlot(reference, 
             reduction.use = "umap", var = "cnv_filtered_geneset_norm_ref", 
             do.label = F, size = 0.2)
```

```r
ggsave('figures/iCNV.pdf', width = 8)
```

```r
options(repr.plot.height = 7, repr.plot.width = 8)
dittoDimPlot(reference, 
             reduction.use = "umap", var = "gender", 
             order = 'decreasing',
             color.panel = c(dittoColors()[1:2], '#e8e8e8'),
             do.label = F, size = 0.2)
```

```r
ggsave('figures/gender.pdf', width = 8)
```
