```r
suppressPackageStartupMessages({
    library(ggplot2)
    library(Seurat)
    library(dittoSeq)
})
```

```r
gbm <- readRDS('data/extended_GBmap.rds')
gbm
```

```r
gbm@meta.data$Tissue<- factor(x = gbm@meta.data$Tissue, 
                                   levels = rev(c('NA', 'Periphery_Xie',
                                             'Tumor_Xie', 'Tumor', 'Periphery')))
```

```r
options(repr.plot.height = 10, repr.plot.width = 10)
dittoDimPlot(gbm, do.raster = TRUE,
             reduction.use = "umap", var = "Tissue", order = 'decreasing',
             color.panel = c('#d55e00','#0072b2','#f0e442','#009e73', '#e5e5e5'),
             do.label = F, labels.size = 3, size = 1)
```
