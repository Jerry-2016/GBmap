```r
suppressPackageStartupMessages({
    library(ggplot2)
    library(Seurat)
    library(dplyr)
    library(dittoSeq)
})
```

```r
gbm <- readRDS('data/Ruiz2021_manual_annotation.rds')
```

```r
options(repr.plot.height = 10, repr.plot.width = 10)
dittoDimPlot(gbm, 
             reduction.use = "umap", var = "orig.ident", 
             do.label = F, size = 1.5) + NoAxes() 
```

```r
ggsave('figures/Ruiz2021_by_patient.pdf', height = 9,
       width = 10)
```

```r
options(repr.plot.height = 10, repr.plot.width = 10)
dittoDimPlot(gbm, 
             reduction.use = "umap", var = "leiden", 
             do.label = F, size = 1.5) + NoAxes() 
```

```r
ggsave('figures/Ruiz2021_leiden.pdf', height = 9,
       width = 10)
```

```r
options(repr.plot.height = 10, repr.plot.width = 10)
dittoDimPlot(gbm, 
             reduction.use = "umap", var = "ident", 
             color.panel = c(c('#d62728', '#ff7f0e', '#3182bd','#31a354',
               '#e6ab02', '#00F6B3',
               '#637939','#8c6d31',
                '#7b4173',
                '#756bb1','#984064')),
            do.raster = TRUE,
            size = 1.5) + NoAxes() 
```

```r
ggsave('figures/Ruiz2021_manual_annotation.pdf', height = 9,
       width = 10)
```

```r
gbm_mapped <- readRDS('data/mapped_Ruiz2021_to_core_GBmap.rds')
```

```r
options(repr.plot.height = 10, repr.plot.width = 10)
gbm$predicted.high_hierarchy <- gbm_mapped@meta.data$predicted.high_hierarchy

gbm@meta.data$predicted.high_hierarchy<- factor(x = gbm@meta.data$predicted.high_hierarchy, 
                                   levels = c('AC-like','MES-like','OPC-like','NPC-like',
     'Astrocyte','Oligodendrocyte','OPC','Neuron',
     'TAM-BDM','TAM-MG','DC',
     'CD4/CD8',
     'Pericyte','Endothelial'))

dittoDimPlot(gbm, 
             reduction.use = "umap", var = "predicted.high_hierarchy", 
            do.raster = TRUE,
            color.panel = c('#d62728', '#ff7f0e', '#3182bd','#31a354',
               '#e7298a','#e6ab02', '#bf5b17','#00F6B3',
               '#637939','#8c6d31','#636363',
                '#7b4173',
                '#756bb1','#17becf'),
            size = 1.5) + NoAxes() 
```

```r
ggsave('figures/Ruiz2021_annotation_high_hierarchy.pdf', height = 9,
       width = 10)
```
