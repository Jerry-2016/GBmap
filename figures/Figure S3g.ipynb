```r
suppressPackageStartupMessages({
    library(ggplot2)
    library(Seurat)
    library(dittoSeq)
})
```

```r
gbm <- readRDS('data/extended_GBmap.rds')
gbm
```

```r
options(repr.plot.height = 30, repr.plot.width = 30)
dittoDimPlot(gbm, split.by = 'author', do.raster = TRUE,
             reduction.use = "umap", var = "author", 
             do.label = F, labels.size = 3, size = 0.5)
```

```r
ggsave('figures/extended_GBmap_split_by_author.pdf', height = 10,
       width = 12)
```
